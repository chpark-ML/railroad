```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 평균과 표준편차 설정
mean = 0
std_dev = 5.0

# 샘플링할 총 개수
total_samples = 1000

# 샘플링
samples = np.abs(np.random.normal(mean, std_dev, total_samples))

# 구간 설정
num_bins = 20
bin_edges = np.linspace(samples.min(), samples.max(), num_bins + 1)
```

```python
# 구간별 샘플 개수 계산
hist, _ = np.histogram(samples, bins=bin_edges)

# 히스토그램 그리기
plt.hist(samples, bins=bin_edges, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Samples from Normal Distribution')
plt.show()
```
