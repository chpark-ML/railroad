```python
import os
from glob import glob

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import projects.DC_prediction.utils.constants as C
import projects.DC_prediction.datasets.railroad as dataset
```

```python
def _plot_line_graph(df: pd.DataFrame, y_col: str, n_col: int, current_col: int):
    plt.subplot(n_col, 1, current_col)
    plt.plot(df.index, df[y_col], linewidth=1, color="r" if y_col in C.PREDICT_COLS else "blue")
    plt.title(f'{y_col}', size="large")
    if current_col != n_col - 1:
        plt.xticks([])  # X 축 눈금 없애기
    plt.ylabel(y_col)
    plt.grid(True)


def plot_line_graph(df_path):
    df = pd.read_csv(df_path)
    columns_to_plot = df.columns
    plt.figure(figsize=(30, 70))
    _counter = 1
    for col in columns_to_plot:
        _plot_line_graph(df, col, len(columns_to_plot), _counter)
        _counter += 1
    plt.tight_layout()
    plt.show()
```

```python
plot_line_graph('result_test.csv')
```
